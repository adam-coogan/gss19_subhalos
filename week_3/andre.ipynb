```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as integrate
from IPython.display import Latex
from scipy.interpolate import interp1d

# comment out the following lines if seaborn is not installed
import seaborn as sns
sns.set(rc={'text.usetex' : True})
sns.set_context("paper", font_scale=2)
sns.set_style("ticks", {'axes.grid': False,
                        'axes.linewidth': 1.5,
                        'legend.frameon':True,
                        'useTex':True})
```

```python
sigmaSI_data = np.genfromtxt('sigmaSI.csv', delimiter=',')
                             
sigmaSI=interp1d(sigmaSI_data[:,0],sigmaSI_data[:,1],kind='linear',fill_value='extrapolate')
WIMP_mass_vect = np.logspace(1, 3)                
plt.figure(figsize=(10,5))
plt.plot(WIMP_mass_vect, sigmaSI(WIMP_mass_vect))
plt.yscale('log')
plt.xscale('log')
plt.ylabel(r'$\sigma_{SI}$ (cm$^2$)')
plt.xlabel(r'WIMP mass (GeV)')
plt.show()
```
